***NumPy Basics***

NumPy is a Python module that we can use instead of built-in arrays and lists. Firstly, the NumPy is faster than built-in arrays and more memory efficient than lists. Additionally, this module includes various methods that can be used in linear algebra, machine learning, data science etc. So, for data science, we must know this module and its methods.

When we want to use the NumPy or any other module, firstly we must **import** it. Before importing, you must download this module via **pip** or **Anaconda** to your Python. After downloading the module, you can import and then use it.

```python
import numpy as np # "np" is a common abbreviation for NumPy, we can use NumPy with this alias.
```

After importing NumPy, we can create a NumPy array with **array()** method. You can create multi-dimensional arrays in NumPy.

```python
new_array = np.array([1, 2, 3, 4, 5]) # 1D array(vector)
new_array_2d = np.array([[1, 2, 3],[4, 5, 6]]) # 2D array(matrix)
print("Vector: ", new_array)
print("Matrix: ", new_array_2d)
```

Also, we can change the shape of our array in NumPy. We can use **shape** attribute to learn the shape of our NumPy array and **reshape()** method to change the shape of array. **Don't forget to assign reshaped array to a variable.**

```python
new_array = np.array([1, 2, 3, 4, 5, 6, 7, 8])
print("Shape: ", new_array.shape)
print(new_array)
```

```python
new_array_reshaped = new_array.reshape((4, 2)) # This method converts the 1D vector to the 2D matrix.
print("The shape of reshaped array", new_array_reshaped.shape)
print(new_array_reshaped)
```

We can learn the size of the NumPy array with **size** attribute, the dimension of the NumPy array with **ndim** attribute, the data type of the array with **dtype.name** attribute. 

```python
print("Size: ", new_array_reshaped.size) # Size of the array(the number of elements)
print("Dimension: ", new_array_reshaped.ndim) # Dimension of the array
print("Data type of array: ", new_array_reshaped.dtype) # What data type the array includes?
```

We can initialize a NumPy array with full of zeros, ones and other values. So, we can use the filled array for future operations. We can create an array full of zeros and ones with **zeros()** and **ones()** method. Also, we can create an array with random values with **empty()** method.

```python
zeros = np.zeros((4, 2)) # We must include the shape of array in arguments. 
print(zeros)
ones = np.ones((4, 2)) # Same as zeros() method.
print(ones)
```

```python
empty = np.empty((5, 5)) # Same as zeros() method.
print(empty)
```

We can fill in the array with definite values, **arange()** method is used for filling the array with sorted values within a range **indicating the step**. **linspace()** method is used for filling the array with **certain number of sorted** values between a start and stop points.

```python
new_new_array = np.arange(10, 50, 5) # Start point, end point(excluded), step
new_new_array
```

```python
new_new_array = np.linspace(10, 50, 20) # Start point, end point(can be inclusive and exclusive), number of elements
new_new_array
```

***Basic Operations***

We can make various mathematical operations for NumPy arrays. "**addition**" with **add()** method, "**subtraction**" with **subtract()** method, "**division**" with **divide()** method, calculate sinus with **sin()** method. All these functions are used "**element-wise**". Only for multiplication, we can use "**element-wise product**" and "**matrix product**". You can check this [link](https://numpy.org/doc/stable/reference/routines.math.html) for all possible mathematical operations on NumPy. Remember that, the shapes of both arrays **must be the same**.

```python
first_array = np.array([[2, 4, 6, 8, 10], [3, 5, 7, 9, 11]])
second_array = np.array([[3, 6, 9, 12, 15], [4, 6, 8, 10, 12]])

print("Addition array: \n", np.add(first_array, second_array)) # Addition
print("Subtraction array: \n", np.subtract(first_array, second_array)) # Subtraction
print("Division array: \n", np.divide(first_array, second_array)) # Division
print("Sinus array: \n", np.sin(first_array)) # Sinus
print("Element-wise production array: \n", np.multiply(first_array, second_array)) # Element-wise product
print("Matrix production array: \n", np.dot(first_array, second_array)) # Matrix product
```

**dot()** method gives an error, because for matrix multiplication, **the number of columns of the first array** and **the number of rows of the second array** must be the same. We can solve this problem with taking the transpose of the second array. Because, it will change the second array's shape to (5,2). Since the shape of the first array is (2, 5), we can make matrix multiplication between these arrays.

```python
second_array_transposed = second_array.T
print(second_array_transposed.shape)
```

```python
print("Matrix production array: \n", np.dot(first_array, second_array_transposed))
```

We can calculate the sum and mean of rows and columns one by one. For this purpose, **sum()** and **mean()** methods have "**axis**" argument. When we specify axis as 0, this means we do the operations **column-wise**. If the value is 1, we do the operations **row-wise**.

```python
print(first_array)
print("Column-wise sum array: " , first_array.sum(axis=0)) # Addition column-wise
print("Row-wise sum array: ", first_array.sum(axis=1)) # Addition row-wise
print("Column-wise mean: ", first_array.mean(axis=0)) # Mean column-wise
print("Row-wise mean: ", first_array.mean(axis=1)) # Mean row-wise
```

*Indexing and Slicing*

We can choose the elements by their indexes in arrays. Indexes can be chosen one by one.

```python
array_1d = np.array([1, 2, 3, 4, 5])
array_2d = np.array([[2, 4, 6, 8, 10], [3, 6, 9, 12, 15]])

print(array_1d[0]) # Select first element of 1D array
print(array_2d[1, 2]) # Select the element in the second row and third column.
```

If we choose a portion of a array, we can use **slicing**. The main rule of slicing is we must give a start and stop point. Stop point is not included. You can mention the "step size", if you want.

```python
print(array_1d[0:3]) # Select from first element to third element, fourth is not included
print(array_2d[:,1]) # Print all rows in the second column
print(array_2d[1, 1:4]) # Print the second row of the portion consisting the second as start point and fourth columns as stop point
# Negative Indexing(starting from the last)
print(array_2d[:, -1]) # Print all rows in the last column
print(array_2d[-1, :]) # Print all columns in the last row
print(array_1d[::-1]) # Reverse an array(Last mentioned -1 is the "step size")
```

*Shape Manipulation*

We can change the shape of the NumPy as our needs. We can change the 2D matrix into 1D vector with **ravel()** method.

```python
print("Flattened array: \n", array_2d.ravel()) # Turn into 1D vector
print("Old shape: \n", array_2d.reshape(2, 5)) # Turn back to old shape
```

We take the transpose of our matrix with "**T**" attribute.

```python
array_2d.T
```

We can also change the shape of our array with **resize()** method, but different from **reshape()**, this method changes the original array.

```python
print("Reshaped version: ", array_2d.reshape((10, )))
print("Old version: ", array_2d) # Come back to old shape
```

```python
array_2d.resize((10, ))
print("Original array: ", array_2d) # Method is changed the original array
```

*Stacking Arrays*

Sometimes, we need to concatenate arrays. Joining can be vertical or horizontal. If we join two arrays in a vertical way, we use **vstack()** function. If we join in a horizontal way, we use **hstack()** function.

```python
array_first = np.array([1, 2, 3, 5, 6])
array_second = np.array([2, 4, 7, 9, 10])

joined_vertical = np.vstack((array_first, array_second)) # Write the arrays in a tuple
joined_horizontal = np.hstack((array_first, array_second))

print("Vertical array: \n", joined_vertical)
print("Horizontal array: ", joined_horizontal)
```

*Converting and Copying Arrays*

If we have a list or tuple and we want to convert into a NumPy array, we can use **array()** function like we use in the "array creation" part. We can use also **asarray()** function to make an array.

```python
list = [1, 2, 3, 4, 5]
converted_one = np.array(list)
print("Type of the first one: ", type(converted_one))

tuple = (1, 2, 3, 4, 5)
converted_two = np.array(tuple)
print("Type of the second one: ", type(converted_two))

converted_back_one = converted_one.tolist() # Convert back to list

print("Type of the first one now: ", type(converted_back_one))
```

When an array is copied with assignment, it can be seen that they are referencing the same array. So, when a change is made in the original array, the copy of the other array is also affected. 

```python
first_array = np.array([1, 2, 3, 4, 5])
second_array = first_array
print("ID of the first array: ", id(first_array)) # id() method is used for detecting the ID of an object
print("ID of the second array: ", id(second_array))
first_array[0] = 45
print("First element of the first array: ", first_array[0])
print("First element of the second array: ", second_array[0]) # They will print the same value
```

How can we copy and create a "**fully separated**" array? For this purpose, **copy()** method is used for copying the array. After copying process, if we change any element of the original array, the other array is not affected. 

```python
first_array = np.array([1, 2, 3, 4, 5])
second_array = first_array.copy()
print("ID of the first array: ", id(first_array))
print("ID of the second array: ", id(second_array))
first_array[0] = 45
print("First element of the first array: ", first_array[0])
print("First element of the second array: ", second_array[0])
```

*Pandas Basics*

Another module used by data scientists is **Pandas**. It uses **DataFrames** to store the data. There are various advantages for using the Pandas. Firstly, it is **fast and effective** with DataFrames. Secondly, we can use **csv and txt** files for **storing and handling** the data. Thirdly, We can also handle the **missing data** in the datasets with this module. The other advantages are, we can reshape easily, more effective indexing and slicing and using with "**time-series data analysis**".

For using this module, This module must be imported first.

```python
import pandas as pd # Universal alias for pandas
```

After that, we can create our own DataFrames from lists, tuples, dictionaries and other data structures with **DataFrame()** method, like NumPy's array() method.

```python
# In this example, we are using dictionary of lists, because the DataFrame is a 2-dimensional object.
dictionary_the_first = {"name": ["Ahmet", "Mehmet", "Çetin", "Güler", "Mine", "Dilek"], "surname": ["Kayapınar", "Bozdoğan", "Tandoğan", "Güner", "Kayıkçı", "Şen"]
                       ,"age": [24, 26, 35, 40, 29, 37], "salary": [5000, 7500, 10000, 12500, 15000, 20000]}
first_DataFrame = pd.DataFrame(dictionary_the_first)
first_DataFrame
```

*Basic Methods*

For inspecting the DataFrames, there are two main methods that show a little portion. These methods are **head(n)** and **tail(n)**. (n is a variable that we assign, it's 5 by default.) head() method shows the first n elements, tail() method shows the last n element in the DataFrame.

```python
print(first_DataFrame.head()) # Shows the first 5 element
print("\n")
print(first_DataFrame.tail()) # Shows the last 5 element
print("\n")
print(first_DataFrame.head(n=3)) # We can assign a new n value
print("\n")
print(first_DataFrame.tail(n=3))
```

For seeing which columns are included in the DataFrame, we can use "**columns**" attribute.

```python
first_DataFrame.columns
```

**info()** method shows the information about the DataFrame. Here are the meanings of every term in the output of this method.
* RangeIndex: How many elements or entries are they in the DataFrame, and how the indexes numerated?
* null/non-null: If there is a missing information, this block is counted as "null", otherwise, the column has non-null values.
* dtypes: What data types do columns consist of? ("string" is named as "object" in here.)

```python
first_DataFrame.info()
```

"**dtypes**" attribute gives the datatypes of the columns.

```python
first_DataFrame.dtypes
```

"**describe()**" method gives statistical information about **numeric** and **categorical** type of columns. Here are the meanings of the terms in the output:
* **count**: How many elements are there in this column?
* **unique**: How many unique elements are in the column?
* **top**: What is the most common value in the category type?
* **freq**: What is the frequency of the most common value in the category type?
* **mean**: Mean of this column
* **std**: Standard deviation of this column
* **min/max**: Minimum and maximum values of the column
* **25%/50%/75%** : First quartile, median and third quartile values of the column

```python
first_DataFrame.describe(include='all') 
# Describe all statistics for categorical and numeric type, 
# it shows only the statistics of numeric type by default. We added an "include" parameter.
```

*Indexing and Slicing*

Like NumPy, for choosing a partition in Pandas, various expressions are used. Let us see the most common ones.

```python
print(first_DataFrame["name"]) # In the square brackets, column name is mentioned. This way can be used to choose a column.
print()
print(first_DataFrame.name) # Another column choosing method.
```

"**loc[]**" method is an abbreviation of "locate" word. We can locate our data with indexing and slicing.

```python
print(first_DataFrame.loc[: ,"age"]) # Whole age column
print("\n")
print(first_DataFrame.loc[: ,"name":"age"]) # Slicing from name to age column(age is included)
print("\n")
print(first_DataFrame.loc[0:2, "age"]) # Ages of the first three people(2 is included).
#  Remember that, 0 and 2 are "labels". Not indexes. For index-based location, we must use another method.
print("\n")
print(first_DataFrame.loc[1:4, :]) # All information about the second to fifth people
print("\n")
print(first_DataFrame.loc[2:4, "surname":"salary"])
```

"**.iloc[]**" method is used as integer-based location. To use this function, users must give a single index or the indexes of the start and stop points in slicing.

```python
print(first_DataFrame.iloc[:, 2]) # Prints the third column
print("\n")
print(first_DataFrame.iloc[2, :]) # Prints the third row
print("\n")
print(first_DataFrame.iloc[0:2, :]) # Prints the first two row(2 is not included)
print("\n")
print(first_DataFrame.iloc[:, 0:2]) # Prints the first two column
print("\n")
print(first_DataFrame.iloc[0:2, 0:3]) # Prints the first three column of the first two rows
```

*Filtering*

DataFrames can be filtered with **boolean expressions**. This provides us to make analysis about our dataset. Here are some examples how we do the analysis:

```python
# We can examine the salaries of the people are greater than 15000 or not.
print(first_DataFrame.salary > 15000) # This method returns boolean Series object.
```

If we add this expression in a square bracket and apply to our dataset, we can filter the dataset with people with salary greater than 15000.

```python
first_DataFrame[first_DataFrame.salary > 15000] # Filter the people who has a salary more than 15000.
```

*List Comprehension*

When we want to apply an expression to each of our samples and create a new feature with it, we can use "list comprehension" which we use **an expression** and a **for loop**.

```python
mean_salary = np.mean(first_DataFrame.salary)
first_DataFrame["salary_level"] = ["low" if each < mean_salary else "high" for each in first_DataFrame.salary]
print("Mean salary: ", mean_salary)
first_DataFrame
```

*Dropping and Concatenating Data*

We can remove a feature with **drop()** method. We can add "**axis**" argument which indicates the removing way is whether vertical or horizontal and "**inplace**" argument which removes the feature permanently.

```python
first_DataFrame.drop(["age"], axis=1, inplace=True) # axis=1 means the removing process is done vertically
print(first_DataFrame) # Age column is deleted permanently
```

We can also append datasets vertically or horizontally with **concat()** method.

```python
data1 = first_DataFrame.head()
data2 = first_DataFrame.tail()
second_DataFrame = pd.concat([data1, data2], axis=0) # axis=0 means connect vertically, 1 means horizontally
print(second_DataFrame)
```

*Transforming Data*

Like "List comprehension", we can apply our own function to the columns with **apply()** method.

```python
def increase_salary(x):
    return x + 1000 # With this function, we can increase salaries
```

```python
series = second_DataFrame.salary.apply(increase_salary)
second_DataFrame["salary"] = series
print(second_DataFrame)
```

***Matplotlib Basics***

Data visualization is a critical part for data science. Graphs or plot are showing the data and we can make inferences from the plots. One of the most used module for data visualization is "**matplotlib**". Here, we see the basic plots in the matplotlib module.

Firstly, like other modules, we must import the matplotlib module, but not the all module. For our purposes, pyplot subclass is enough.

```python
import matplotlib.pyplot as plt # Universal alias for matplotlib module
```

*Line Plot*

Line plots are used for showing the changes by time or showing the anomalies in data. In matplotlib, when the **plot()** method is called to draw a plot without any adjustment in the arguments, this type of plot is drawn by default This situation makes the line plot one of the most used chart types.

In this section, we're going to use "**iris**" dataset to show plots. You can download the dataset from this [link](https://archive.ics.uci.edu/ml/datasets/iris).

```python
iris = pd.read_csv("../input/iris/Iris.csv") # First, we must import our dataset to interpreter
# After importing the data, we must analyze the data with statistics. We will do this process with these methods below.
iris.info() # We can see the data types and number of entries here
iris.describe(include='all') #  We can see statistics of the numeric data
```

With this dataset, our primary objective is classification of the flowers from their length and width of the sepal and petal parts. We see that standard deviation of the "**Petal Length**" is more than the other attributes, so we can classify these three species with it. Let us see first the name of species.

```python
iris["Species"].unique()
```

Then, we seperate these flowers into different dataframes by their species.

```python
setosa = iris[iris["Species"] == "Iris-setosa"]
versicolor = iris[iris["Species"] == "Iris-versicolor"]
virginica = iris[iris["Species"] == "Iris-virginica"]

print(setosa.head())
print(versicolor.head())
print(virginica.head())
```

After that, we can show the distribution of the petal lengths via line plot of matplotlib module. We use "**plot()**" method to draw line plot. We specify the **x and y axes** and the **color of each line**. Additionally, for easy reading, we can add **labels** to our plot, don't forget to add legend with "**legend()**" method to see which label is used for which kind.

```python
plt.plot(setosa["Id"], setosa["PetalLengthCm"], color="red", label="setosa")
plt.plot(versicolor["Id"], versicolor["PetalLengthCm"], color="blue", label="versicolor")
plt.plot(virginica["Id"], virginica["PetalLengthCm"], color="green", label="virginica")
plt.legend()
plt.show()
```

Seems like, we can classify setosas easily with their petal lengths **according to our samples**. On the other hand, we can not seperate versicolor and virginica **completely** with this information.

Next, we can beautify this plot with adding "x-labels" and "y-labels" and a "title".

```python
plt.plot(setosa["Id"], setosa["PetalLengthCm"], color="red", label="setosa")
plt.plot(versicolor["Id"], versicolor["PetalLengthCm"], color="blue", label="versicolor")
plt.plot(virginica["Id"], virginica["PetalLengthCm"], color="green", label="virginica")
plt.xlabel("Id of flower")
plt.ylabel("Petal Length(cm)")
plt.title("Petal Lengths of iris flowers by species")
plt.legend()
plt.show()
```

We can also change the style of lines with "**linestyle**" attribute of the plot() method.

```python
plt.plot(setosa["Id"], setosa["PetalLengthCm"], color="red", label="setosa", linestyle=":") # "dotted" linestyle
plt.plot(versicolor["Id"], versicolor["PetalLengthCm"], color="blue", label="versicolor", linestyle="-.") # "dashdot" linestyle
plt.plot(virginica["Id"], virginica["PetalLengthCm"], color="green", label="virginica", linestyle="--") # "dashed" linestyle
plt.xlabel("Id")
plt.ylabel("Petal Length(cm)")
plt.title("Petal Lengths of iris flowers by species")
plt.legend()
plt.show()
```

We can add grid lines with **grid()** method before showing the data. Also, we can add "**alpha**" attribute to adjust transparency of the line. It takes a value between 0 and 1. The closer it gets to zero, the more transparent line becomes.

```python
plt.plot(setosa["Id"], setosa["PetalLengthCm"], color="red", label="setosa", linestyle=":", alpha=0.1) # Nearly full transparent
plt.plot(versicolor["Id"], versicolor["PetalLengthCm"], color="blue", label="versicolor", linestyle="-.", alpha=0.5) # "Half transparent
plt.plot(virginica["Id"], virginica["PetalLengthCm"], color="green", label="virginica", linestyle="--", alpha=0.9) # Nearly full opaque
plt.xlabel("Id")
plt.ylabel("Petal Length(cm)")
plt.title("Petal Lengths of iris flowers by species")
plt.grid()
plt.legend()
plt.show()
```

*Scatter Plot*

What if we examine the relationship between two features? How can we show the relation with plot? **Scatter plot** can be an answer to these questions. We use "**scatter()**" method to draw a scatter plot. We're going to draw a scatter plot from petal length and petal width of the flowers below.

```python
plt.scatter(setosa["PetalLengthCm"], setosa["PetalWidthCm"], color="red", label="setosa")
plt.scatter(versicolor["PetalLengthCm"], versicolor["PetalWidthCm"], color="blue", label="versicolor")
plt.scatter(virginica["PetalLengthCm"], virginica["PetalWidthCm"], color="green", label="virginica")

plt.xlabel("Petal Length(cm)")
plt.ylabel("Petal Width(cm)")
plt.title("Scatter Plot")
plt.legend()
plt.show()
```

We can infer that we can also classify species with petal width in **Setosas**, but still we can not separate **Versicolor and Virginica** between each other with these features.

We can add "**alpha**" attribute to adjust transparency of dots. 

```python
plt.scatter(setosa["PetalLengthCm"], setosa["PetalWidthCm"], color="red", label="setosa", alpha = 0.2)
plt.scatter(versicolor["PetalLengthCm"], versicolor["PetalWidthCm"], color="blue", label="versicolor", alpha=0.5)
plt.scatter(virginica["PetalLengthCm"], virginica["PetalWidthCm"], color="green", label="virginica", alpha=0.9)

plt.xlabel("Petal Length(cm)")
plt.ylabel("Petal Width(cm)")
plt.title("Scatter Plot")
plt.legend()
plt.show()
```

*Histogram*

When our aim is to show distribution of a feature in the data, we use histograms. For drawing histograms, we use "**hist()**" method. We can add "**bins**" attribute to specify the number of ranges between the minimum and maximum value of the feature. Default value for "bins" is 10. Here we can see the distribution of the values in petal lengths of setosas. 

```python
plt.hist(setosa["PetalLengthCm"]) # We create 10 bins to include values
plt.xlabel("Petal Length(cm)")
plt.ylabel("Frequency of the Value")
plt.title("Histogram")
plt.legend()
plt.show()
```

```python
plt.hist(setosa["PetalLengthCm"], bins=3) # Now, we create 50 bins to include values
plt.xlabel("Petal Length(cm)")
plt.ylabel("Frequency of the Value")
plt.title("Histogram")
plt.legend()
plt.show()
```

Here, we can see that more bins mean **less number of values in each of bin**. So, we can examine more effectively, but we must test the bin sizes to find the perfect size of bins.

*Bar Plot*

When we make comparison between groups or species, we can use this graph. We use "**bar()**" method for drawing a bar plot. It takes two arrays or series, in the x-axis, we mention the groups, in y-axis, we mention values of them. We compare the mean values of petal lengths to make an inference about species.

```python
plt.bar(["Setosa", "Versicolor", "Virginica"],[np.mean(setosa["PetalLengthCm"]), np.mean(versicolor["PetalLengthCm"]), np.mean(virginica["PetalLengthCm"])])
plt.xlabel("Species")
plt.ylabel("Petal Length(cm)")
plt.title("Bar Plot")
plt.show()
```

We see from the graph that, the petal lengths of Virginica is taller than Versicolor. Means are close to each other, so we can not make a **complete inference** from petal lengths in these species.

*Subplots*

Sometimes, one graph is **not enough** to compare values or we can see the differences more easily when we see two, three or more graphs in one plot. For this situations, we create subplots with "**subplot()**" method. We specify **number of rows and columns** and the **location of the graph** as arguments.

```python
plt.subplot(2, 1, 1) # (number_of_rows, number_of_columns, index)
plt.plot(setosa["PetalLengthCm"], color="red")
plt.ylabel("Setosa - Petal Length(cm)")
plt.subplot(2, 1, 2)
plt.plot(versicolor["PetalLengthCm"], color="blue")
plt.ylabel("Versicolor - Petal Length(cm)")
plt.show()
```

We can see from y-axis values that Versicolors' petal lengths are taller than Setosas'.
